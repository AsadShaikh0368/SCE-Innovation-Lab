### Plotting a Dynamic Cube given input face N
Plot a __N X N X N__ cube with <strong>N<sup>3</sup></strong> cubes inside

for e.g.
```
input N = 3
output:
Here's a 3 X 3 X 3 cube plots with 27 cubes inside
![n_3](/assets/images/n_3_cube.png "3-sided cube")

input N = 4
output:
Here's a 4 X 4 X 4 cube plots with 64 cubes inside
![N_4_cube](/assets/images/n_4_cube.png)

input N = 7
output:
Here's a 7 X 7 X 7 cube plots with 343 cubes inside
![N_7_cube](/assets/images/n_7_cube.png)

```

````python
#import itertools
import matplotlib.pyplot as plt, io
from operator import concat

### figure properties:
n_dimension = 3
plot_range = list(range(n_dimension + 1))
ax = plt.figure().add_subplot(111, projection='3d')

# using reflection to set three variables for three axes instead of doing it hard coded
axes_list = [['x', 'X'], ['y', 'Y'], ['z', 'Z']]
figure_properties = ['label', 'lim', 'ticks']
set_label, set_lim, set_ticks = tuple(map(concat, ('set_{co_ord}', )*3, figure_properties))
for co_ordinate in axes_list:
    #print(co_ord[0], co_ord[1])
    getattr(ax, set_label.format(co_ord = co_ordinate[0]))(co_ordinate[1])
    getattr(ax, set_lim.format(co_ord = co_ordinate[0]))([0, n_dimension])
    getattr(ax, set_ticks.format(co_ord = co_ordinate[0]))(plot_range)

### PLOTTING THE CUBE ###
# dynamically tracing cube lines
for i in plot_range:
    # plotting the outside large faces
    # x facing plane
    ax.plot([0,n_dimension], [i,i], [0,0], color='black')
    ax.plot([0,0], [i,i], [0,n_dimension], color='black')
    ax.plot([0,n_dimension], [i,i], [n_dimension,n_dimension], color='black')
    ax.plot([n_dimension,n_dimension], [i,i], [0,n_dimension], color='black')
    # y facing plane
    ax.plot([i,i], [0,n_dimension], [0,0], color='black')
    ax.plot([i,i], [0,0], [0,n_dimension], color='black')
    ax.plot([i,i], [0,n_dimension], [n_dimension,n_dimension], color='black')
    ax.plot([i,i], [n_dimension,n_dimension], [0,n_dimension], color='black')
    # z facing plane
    ax.plot([0,n_dimension], [0,0], [i,i], color='black')
    ax.plot([0,0], [0,n_dimension], [i,i], color='black')
    ax.plot([0,n_dimension], [n_dimension,n_dimension], [i,i], color='black')
    ax.plot([n_dimension,n_dimension], [0,n_dimension], [i,i], color='black')
    # plotting internal lines
    for j in range(1, len(plot_range) -1 ):
        if i < len(plot_range)-1:
            # plotting x facing lines
            ax.plot([0,n_dimension], [j,j], [i+1,i+1], color='black')
            # plotting y facing lines
            ax.plot([j,j], [0,n_dimension], [i+1,i+1], color='black')
    
print("Here's a {N} X {N} X {N} cube plots with {N_raise_to_power_3} cubes inside"
      .format(
          N=n_dimension,
          N_raise_to_power_3=(n_dimension**3)
      )
     )
plt.show()
````
